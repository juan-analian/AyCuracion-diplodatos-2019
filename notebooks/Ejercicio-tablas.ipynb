# Práctico 1
Graficar las curvas correspondientes a los percentiles, separando en 2 graficos los niños y las niñas.

```python
plt.style.use('bmh')
tabla_1[tabla_1.Sex == 1].plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'], figsize=(10,5), grid=True)
plt.title("Niños: Percentiles segun meses de vida")
plt.ylabel("Peso")
plt.xlabel("Edad en meses")
plt.show()
```

```python
plt.style.use('bmh')
tabla_1[tabla_1.Sex == 2].plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'], figsize=(10,5), grid=True)
plt.title("Niñas: Percentiles segun meses de vida")
plt.ylabel("Peso")
plt.xlabel("Edad en meses")
plt.show()
```
